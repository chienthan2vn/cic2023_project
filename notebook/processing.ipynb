```python
# Get counts of each data type in the DataFrame
dtype_counts = df.dtypes.value_counts()
print("Data type distribution:")
print(dtype_counts)
print("\nColumns by data type:")
for dtype in dtype_counts.index:
    print(f"\n{dtype}:")
    print(df.select_dtypes(include=[dtype]).columns.tolist())
```

```python
def preprocess_dataframe(df):
    from sklearn.preprocessing import RobustScaler, LabelEncoder, StandardScaler
    df_copy = df.copy()

    # Apply frequency encoding to 'protocol_type'
    protocol_type_counts = df_copy['protocol_type'].value_counts()
    df_copy['protocol_type'] = df_copy['protocol_type'].map(protocol_type_counts)

    # Apply RobustScaler to continuous data
    continuous_columns = df_copy.select_dtypes(include=[np.number]).columns
    scaler = RobustScaler()
    df_copy[continuous_columns] = scaler.fit_transform(df_copy[continuous_columns])

    # Binary encode boolean data
    bool_columns = df_copy.select_dtypes(include=['bool']).columns
    encoder = LabelEncoder()
    for col in bool_columns:
        df_copy[col] = encoder.fit_transform(df_copy[col])

    # Normalize all features except 'label' using StandardScaler
    feature_columns = [col for col in df_copy.columns if col != 'label']
    standard_scaler = StandardScaler()
    df_copy[feature_columns] = standard_scaler.fit_transform(df_copy[feature_columns])

    return df_copy
```

```python
df_processed = preprocess_dataframe(df)
```

```python
# Check if protocol_type was successfully encoded
protocol_cols = [col for col in df_processed.columns if col.startswith('protocol_')]
print(f"Protocol type columns after encoding: {protocol_cols}")

# Display the first few rows of the processed dataset with protocol columns
df_processed[protocol_cols].head()
```

## Protocol Type Processing

The `protocol_type` column was successfully converted from a categorical feature to numeric values using one-hot encoding. Each unique protocol (TCP, UDP, etc.) now has its own binary column, which allows the model to properly understand this categorical feature.
